```python
xm = np.zeros((4401, 33))
ym = np.zeros((4401, 33))
with open('spectre-puzzle-patch-convhull.tsv', 'rb') as f:
    i = 0
    for line in f:
        j = 0
        for point in line.rstrip().split(b'\t'):
            xm[i, j], ym[i, j] = (float(n) for n in point.split(b','))
            j += 1
        i += 1
```

```python
XMIN = -(210/2 - 3)
XMAX = -XMIN
YMIN = -(297/2 - 3)
YMAX = -YMIN
```

```python
def count(dx, dy, r):
    xm2 = xm + dx
    ym2 = ym + dy
    s = np.sin(r)
    c = np.cos(r)
    xm2, ym2 = c*xm2 - s*ym2, s*xm2 + c*ym2
    return np.all(np.logical_and(np.logical_and(XMIN <= xm2, xm2 <= XMAX), np.logical_and(YMIN <= ym2, ym2 <= YMAX)), 1).nonzero()[0]
```
